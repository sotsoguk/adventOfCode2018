```python
import re
f = open('input.txt')
lines = f.readlines()
lines = [l.strip() for l in lines]
robots = []
```

```python
for i,l in enumerate(lines):
    x,y,z,r = list(map(int, re.findall(r'-?\d+',l)))
    robots.append([i,x,y,z,r])
```

```python
robots
```

```python
maxr = lambda x: x[4]
```

```python
maxNano = max(robots,key=maxr)
```

```python
maxId, maxX, maxY, maxZ, maxR = maxNano
```

```python
def manNorm(x1,y1,z1,x2,y2,z2):
    return abs(x1-x2)+abs(y1-y2)+abs(z1-z2)
```

```python
print(maxNano,maxR)
inRange = 0
```

```python
for r in robots:
    if manNorm(r[1],r[2],r[3],maxX,maxY,maxZ) <= maxR:
        inRange += 1
print(inRange)
```
